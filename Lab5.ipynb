```python
import numpy as np

X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 1, 1, 0])

print(X)
print(y)
```

```python
def sigmoid(x):
  return 1.0 / (1.0 + np.exp(-x))
```

```python
def sigmoid_derivative(sigma):
  return sigma * (1.0 - sigma)
```

```python
def MSE(y_predict, y_label):
  return 0.5 * (y_predict - y_label) ** 2
```

```python
def MSE_derivative(y_predict, y_label):
  return y_predict - y_label
```

```python
def train(X, y, epochs, alpha):
  X = np.insert(X, 0, np.ones(len(X)), axis=1)
  w1 = np.random.randn(3, 3)
  w2 = np.random.randn(4)
  sigma1 = np.zeros(4)
  sigma1[3] = 1.0
  for i in range(epochs):
      error = 0.0
      for x_el, y_el in zip(X, y):
        sum1 = np.dot(x_el, w1)
        sigma1[:3] = sigmoid(sum1)
        sum2 = np.dot(sigma1, w2)
        y_pred = sigmoid(sum2)

        error += MSE(y_pred, y_el)

        loss_w2 = MSE_derivative(y_pred, y_el) * sigmoid_derivative(sigmoid(sum2)) * sigma1
        w2 -= alpha * loss_w2

        loss_w1 = MSE_derivative(y_pred, y_el) * sigmoid_derivative(sigmoid(sum2)) * w2[:3] * sigmoid_derivative(sigma1[:3]) * x_el
        w1 -= alpha * loss_w1
      error /= len(y)
      print('Epoch ' + str(i + 1) + ' Error: ' + str(error))
```

```python
epochs = 4000
alpha = 0.1
train(X, y, epochs, alpha)
```
