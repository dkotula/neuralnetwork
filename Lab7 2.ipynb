```python
import matplotlib.pyplot as plt
import numpy as np

def test1():
  x = np.random.normal(0, 1, size=(500, 1))
  z = [x]
  for layer in range(10):
    W = np.zeros((500, 500))
    z.append(np.tanh(np.dot(W, z[-1])))
  f,a = plt.subplots(1, 10, figsize=(25, 2))
  a = a.ravel()
  print('weights to zero')
  for idx,ax in enumerate(a):
      ax.hist(z[idx], bins=[-1, -0.8, -0.6, -0.4, -0.2, 0, 0.2, 0.4, 0.6, 0.8, 1])
      ax.set_title('Layer' + str(idx + 1))
      ax.set_xlim([-1, 1.2])
  plt.show()
  
def test2():
  x = np.random.normal(0, 1, size=(500, 1))
  z = [x]
  print('weights to small random values')
  for layer in range(10):
    W = np.random.normal(0, 0.01, size=(500, 500))
    z.append(np.tanh(np.dot(W, z[-1])))
  f,a = plt.subplots(1, 10, figsize=(25, 2))
  a = a.ravel()
  for idx,ax in enumerate(a):
      ax.hist(z[idx], bins=[-1, -0.8, -0.6, -0.4, -0.2, 0, 0.2, 0.4, 0.6, 0.8, 1])
      ax.set_title('Layer' + str(idx + 1))
      ax.set_xlim([-1, 1.2])
  plt.show()
  
def test3():
  x = np.random.normal(0, 1, size=(500, 1))
  z = [x]
  print('weights to big random values')
  for layer in range(10):
    W = np.random.normal(0, 100, size=(500, 500))
    z.append(np.tanh(np.dot(W, z[-1])))
  f,a = plt.subplots(1, 10, figsize=(25, 2))
  a = a.ravel()
  for idx,ax in enumerate(a):
      ax.hist(z[idx], bins=[-1, -0.8, -0.6, -0.4, -0.2, 0, 0.2, 0.4, 0.6, 0.8, 1])
      ax.set_title('Layer' + str(idx + 1))
      ax.set_xlim([-1, 1.2])
  plt.show()
  
def test4():
  x = np.random.normal(0, 1, size=(500, 1))
  z = [x]
  print('weights with Xavier initialization')
  for layer in range(10):
    W = np.random.normal(0, 1 / np.sqrt(500), size=(500, 500))
    z.append(np.tanh(np.dot(W, z[-1])))
  f,a = plt.subplots(1, 10, figsize=(25, 2))
  a = a.ravel()
  for idx,ax in enumerate(a):
      ax.hist(z[idx], bins=[-1, -0.8, -0.6, -0.4, -0.2, 0, 0.2, 0.4, 0.6, 0.8, 1])
      ax.set_title('Layer' + str(idx + 1))
      ax.set_xlim([-1, 1.2])
  plt.show()
```

```python
test1()
test2()
test3()
test4()
```
