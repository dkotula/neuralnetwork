```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelBinarizer
```

```python
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)
```

```python
print(X_train, y_train)
```

```python
X_train = (X_train - X_train.min(axis=0)) / (X_train.max(axis=0) - X_train.min(axis=0))
X_test = (X_test - X_test.min(axis=0)) / (X_test.max(axis=0) - X_test.min(axis=0))
```

```python
from sklearn.preprocessing import OneHotEncoder

onehot_encoder = OneHotEncoder(sparse=False)
y_train = y_train.reshape(len(y_train), 1)
y_train = onehot_encoder.fit_transform(y_train)
print(y_train)
```

```python
onehot_encoder = OneHotEncoder(sparse=False)
y_test = y_test.reshape(len(y_test), 1)
y_test = onehot_encoder.fit_transform(y_test)
print(y_test)
```

```python
def softmax(x):
  return np.exp(x) / np.sum(np.exp(x))
```

```python
def loss_function(y_predict, y_label):
  return np.sum(-y_label*np.log(y_predict), axis=0)
```

```python
def loss_function_derivative(y_predict, y_label, x):
  return np.matmul(x.reshape((len(x), 1)), (y_predict - y_label.reshape((1, len(y_predict)))))
```

```python
def predict(x, weights):
  return softmax(np.dot(x, weights))
```

```python
def batch_gradient_descent(X, y, W): # update for each epoch
  error = 0.0
  loss = 0.0
  for x_el, y_el in zip(X, y):
    y_pred = predict(x_el, W)
    error += loss_function(y_pred, y_el)
    loss += loss_function_derivative(y_pred, y_el, x_el)
  W -= alpha / len(X) * loss
  error /= len(y)
  return W, error
```

```python
def stochastic_gradient_descent(X, y, W): # update for each iteration
  error = 0.0
  for x_el, y_el in zip(X, y):
    y_pred = predict(x_el, W)
    error += loss_function(y_pred, y_el)
    loss = loss_function_derivative(y_pred, y_el, x_el)
    W -= alpha * loss
  error /= len(y)
  return W, error
```

```python
def mini_batch_gradient_descent(X, y, W, n): # update for n iterations
  error = 0.0
  loss = 0.0
  iter = 0
  for x_el, y_el in zip(X, y):
    iter += 1
    y_pred = predict(x_el, W)
    error += loss_function(y_pred, y_el)
    loss += loss_function_derivative(y_pred, y_el, x_el)
    if iter % n == 0:
      W -= alpha / n * loss
      loss = 0.0
  error /= len(y)
  return W, error
```

```python
def train(X, y, epochs, alpha, gradient_descent, mini_batch = 10): # gradient_descent (1-batch, 2-stochastic, 3-mini_batch)
  input_size = 4
  output_size = 3
  X = np.insert(X, 0, np.ones(len(X)), axis=1)
  W = np.random.randn(input_size + 1, output_size)
  for i in range(epochs):
    if gradient_descent == 1:
      W, error = batch_gradient_descent(X, y, W)
    elif gradient_descent == 2:
      W, error = stochastic_gradient_descent(X, y, W)
    elif gradient_descent == 3:
      W, error = mini_batch_gradient_descent(X, y, W, mini_batch)
    print('Epoch ' + str(i + 1) + ' Error: ' + str(error))
  return W
```

```python
epochs = 2000
alpha = 0.5
W = train(np.array(X_train), np.array(y_train), epochs, alpha, 1)

error = 0.0
for x_el, y_el in zip(np.insert(np.array(X_test), 0, np.ones(len(np.array(X_test))), axis=1), np.array(y_test)):
  y_pred = predict(x_el, W)
  error += loss_function(y_pred, y_el)
error /= len(np.array(y_test))
print('Error over test set: ' + str(error))
```
